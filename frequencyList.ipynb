```python
import pandas as pd
df = pd.read_csv("ContainsDataset.csv")

letters = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j", "k", "l", "m", "n", "o", "p", "q", "r", "s", "t", "u", "v", "w", "x", "y", "z", "ï"]
word_counts = dict()
total_letters = 0
for a in letters:
    word_counts[a] = 0
for word in df['Word']:
    for letter in word: 
        total_letters += 1
        word_counts[letter] +=1
total_letters
```

```python
for i in word_counts:
    word_counts[i] = word_counts[i]/total_letters
```

```python
df["ProbabilityLetter1"] = [word_counts[x[0]] for x in df['Word']]
df["ProbabilityLetter2"] = [word_counts[x[1]] for x in df['Word']]
df["ProbabilityLetter3"] = [word_counts[x[2]] for x in df['Word']]
df["ProbabilityLetter4"] = [word_counts[x[3]] for x in df['Word']]
df["ProbabilityLetter5"] = [word_counts[x[4]] for x in df['Word']]
df
```

```python
df.to_csv("FrequencyWordle.csv")
```
